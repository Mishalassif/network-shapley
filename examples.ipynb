```python
import networkx as nx
import matplotlib.pyplot as plt

s = 100
G = nx.Graph()

G.add_nodes_from([i for i in range(s)])
f = [1 for i in range(s)]
G.add_edge(0, 1)
for i in range(2, s):
    if i%2 == 0:
      G.add_edge(0, i)
    else:
      G.add_edge(1,i)

nx.draw(G, with_labels=True, font_weight='bold')
val = value(G,G)
print('Total value: ' + str(val))

print('Share of major players: ' + str(100*2*shapley(G, 0)/val))
print('Share of minor players: ' + str(100*(len(G.nodes)-2)*shapley(G, 2)/val))
```
